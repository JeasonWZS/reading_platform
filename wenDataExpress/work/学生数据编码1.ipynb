```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_excel('./data/学生个性化数据（修）.xlsx')
df.columns
```

```python
df1 = df[['偏好主要影响品格', '偏好图书内容主题', '偏好图书类型', '偏好图书年代', '偏好作者国别', '主要特别关注',
         '特定定向水平', '行为—认知参考', '感知到的父母关系', '同伴交往主动性', '个体家庭亲密度', '平静、放松',
         '全面发展—智', '全面发展—德', '持有的能力可塑信念', '心理韧性']]
df1 = df1.reset_index()
df1 
```

```python
df1['index'] = df1['index'] + 1
cols = df1.columns.tolist()
cols[0] = 'userid'
cols 
df1.columns = cols
df1 
```

```python
columns_to_encode = ['偏好主要影响品格', '偏好图书内容主题', '偏好图书类型', '偏好图书年代', '偏好作者国别', '主要特别关注']
```

```python
# 初始化映射规则的字典
mapping_rules = {}

# 对每一列进行唯一值映射
for column in columns_to_encode:
    # 使用 factorize 函数进行映射
    df1[column], mapping = pd.factorize(df1[column])
    mapping_rules[column] = {str(old_value): new_value for old_value, new_value in zip(df1[column].unique(), mapping)}

# 打印映射后的 DataFrame 和映射规则
print("Mapped DataFrame:")
print(df1)
print("\nMapping Rules:")
print(mapping_rules)
```

```python
df1.to_excel('./data/学生数据编码1.xlsx', index=False)
```
